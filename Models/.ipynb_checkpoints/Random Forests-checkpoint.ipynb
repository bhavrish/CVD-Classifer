```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# read in data
df = pd.read_csv('../Data/Cluster0', delimiter=',')
print(df.head())
```

```python
df_feat = df.iloc[:,2:-1] # eventhing but last column
y = df.iloc[:,-1] #last column

df_feat.head() # age is in days
```

```python
from sklearn.preprocessing import StandardScaler
# standardize: mean = 0, std = 1
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_feat)
X = pd.DataFrame(scaled_features)
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


for i in range(1,400):
    classifier = RandomForestClassifier(n_estimators=i, oob_score=True)
    classifier.fit(X_train, y_train)
    
    y_pred = classifier.predict(X_test)
    print("i: ", i, ";\tAccuracy:", accuracy_score(y_test, y_pred))
#     print('Score: ', classifier.score(X_train, y_train))
    print(classifier.oob_score_)
    print('---')
```

```python
grid_params = {
    "n_estimators" : range(100, 300, 10),
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [1, 2, 3],
    'min_samples_leaf': [1, 2, 3]
}
```

```python
from sklearn.model_selection import GridSearchCV

RF = RandomForestClassifier(oob_score=True)
gs = GridSearchCV(RF, grid_params, verbose= 1, n_jobs= -1)
gs.fit(X_train, y_train)
gs.best_params_
```

```python
clusters = ["Cluster0", "Cluster1", "Cluster2", "Cluster3", "Cluster4", "Cluster5"]

accuracyDict = {}
bestParamsDict = {}

for clust in clusters:
    # Preprocessing
    path_k = "../Data/" + clust
    df = pd.read_csv(path_k, delimiter=',')
    df_feat = df.iloc[:,2:-1]
    y = df.iloc[:,-1]
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(df_feat)
    X = pd.DataFrame(scaled_features)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
    
    # Train model
    RF = RandomForestClassifier(oob_score=True)
    gs = GridSearchCV(RF, grid_params, verbose= 1, n_jobs=-1)
    gs.fit(X_train, y_train)
#     RF = RandomForestClassifier(n_estimators = gs.best_params_["n_estimators"])
#     RF.fit(X_train, y_train)
    
    # Store accuracies
    y_pred = gs.predict(X_test)
    accuracyDict[clust] = accuracy_score(y_test, y_pred)
    
    # Store best params
    bestParamsDict[clust] = gs.best_params_
    
print(accuracyDict)
print(bestParamsDict)
```

```python
# Entire dataset

df = pd.read_csv("../data/cardio_train.csv", delimiter = ",")
df.head()
```

```python
targ = df['cardio']
```

```python
df_feat = df_feat = df.iloc[:,2:-1]
df_feat.head()
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df_feat)
X = pd.DataFrame(scaled_features)
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, targ)
```

```python
forest = RandomForestClassifier()
gs = GridSearchCV(forest, grid_params, verbose= 2)
gs.fit(X_train, y_train)
forest = RandomForestClassifier(n_estimators=gs.best_params_["n_estimators"])
forest.fit(X_train, y_train)
preds = forest.predict(X_test)
accuracy = accuracy_score(y_test, preds)
accuracy
```

```python
# Save models

from sklearn.pipeline import make_pipeline
import pickle

rf_best_clusters = [2, 5]
best_estimators = [160, 200]

pos = 0

for k in rf_best_clusters:
    inputPath = "../Data/Cluster" + str(k)
    outputPath = "../Web App/Final Models/Cluster_" + str(k) + "_RF_pipeline.pkl"
    
    df = pd.read_csv(inputPath, delimiter=',')
    df = df.iloc[:,2:] # drop original/new ids

    X = df.iloc[:,:-1] # everything but last column
    y = df.iloc[:,-1] #last column

    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(X)
    X = pd.DataFrame(scaled_features)

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    forest = RandomForestClassifier(n_estimators=best_estimators[pos], max_features='log2', min_samples_leaf=3, min_samples_split=2)
    forest.fit(X_train, y_train)
    
    # Create pipeline
    pipeline = make_pipeline(scaler, forest)
    
    # Write model to file
    pickle.dump(pipeline, open(outputPath, 'wb')) 
```
